# 한국 년도별 작품 수

<a href="https://comic.naver.com/webtoon/period.nhn?period=2018">사이트</a>

```python
import time
from PIL import Image
import requests
from io import BytesIO
from selenium import webdriver as wd
# request.get 요청
import pandas as pd
import re
import matplotlib.pyplot as plt
import numpy as np
driver=wd.Chrome("chromedriver")#,options=options)
#https://comic.naver.com/webtoon/period.nhn?period=2018
url="https://comic.naver.com/webtoon/period.nhn?period=2018"
driver.get(url)
time.sleep(1)
cnts = []
yearss= []
#년도별 클릭 li개수 추출
years = driver.find_elements_by_xpath('/html/body/div/div[3]/div[1]/ul/li')
years = len(years)

#그 개수만큼 for문 반복
for i in range(years):
    
    #년도 클릭
    driver.find_element_by_css_selector('#content > ul > li:nth-child({0}) > a'.format(i+1)).click()
    
    
    #해당 년도 값 가져오기
    get_year = driver.find_element_by_css_selector('#content > ul > li:nth-child({0}) > a'.format(i+1))
    #해당 년도에 대한 작품수 추출
    cnt = driver.find_element_by_xpath('/html/body/div/div[3]/div[1]/div[1]/h3/span')
    cnts.append(cnt.text)
    yearss.append(get_year.text)

#['(3)', '(446)', ' (414)', ]cnts는 가공되지 않는 데이터 리스트이다
print(cnts)

#cnts리스트 안에 데이터를 가공
piece = list(map(lambda x: int(re.findall('\d+', x)[0]), cnts))
years = list(map(int, yearss))

#데이터 프레임 생성
df = pd.DataFrame({'년도':years, '작품수' : piece})

data3=piece
#년도가 내림차 순으로 정렬되어 있기 때문에 reverse()함수 사용
data3.reverse()
N = len(data3)
#x축 개수
ind2 = np.arange(N)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

#y축 범위
ax.set_ylim([0,500])
#년도가 내림차 순으로 정렬되어 있기 때문에 reverse()함수 사용
yearss.reverse()
plt.xticks(ind2,yearss) 

for label in ax.xaxis.get_ticklabels() :
    label.set_rotation(45)

    
#한국어 세팅
plt.rcParams["font.family"] = "Malgun Gothic"
plt.title("연도별 국내 웹툰 총 작품수")

#꺽은선 그래프의 ^화살표를 출력
plt.plot(data3,'--^', 0.5)

#x축 년도 문자열 각도를 바꿈
plt.subplots_adjust(bottom=0.1)

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

s1 = [556,559,605,671,692,683,674,663,651,640]
s2 = [478,499,489,485,481,469,457,443,429,413]
s3 = [495,536,540,536,530,519,508,495,481,466]

N = len(s1)
ind = np.arange(N) 

#바의 두께를 세팅
width = 0.23


#-0.25, +0.25에 대한 조정을 하지 않으면 
#바가 겹쳐서 출력 됨
p1 = plt.bar(ind-0.25, s1, width) 
#0을 기준으로 함
p2 = plt.bar(ind, s2, width) 
p3 = plt.bar(ind+0.25, s3, width) 

plt.ylabel('Scores') 
#한국어 세팅
plt.rcParams["font.family"] = "Malgun Gothic"
plt.title('인쇄 만화 시장 규모') 

#ind는 x축 개수만큼 배열을 생성한다
#생성한 배열의 이름을 년도로 변환
plt.xticks(ind, ('2010','2011','2012','2013','2014','2015','2016','2017','2018','2019')) 
plt.rcParams["font.family"] = "Malgun Gothic"

#p1는 자료형이 리스트이며 바에 대한 길이 정보, 수치 값, 색 등의 포함
#p1[0]으로 하나의 바를 선택하고 이름을 지정
plt.legend((p1[0], p2[0],p3[0]), ('미국', '프랑스','독일'))
plt.show()
```
